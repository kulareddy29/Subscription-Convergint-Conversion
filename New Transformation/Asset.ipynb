```python
asset_report_df = pd.read_excel(r"C:\Users\kulareddy\Downloads\Asset Report (1).xlsx")
asset_report_df.head()
```

```python
def normalize_key(series):
    s = series.astype("string")

    s = (
        s.str.replace("\u00A0", "", regex=False)
         .str.replace(r"[\u200B-\u200D\uFEFF]", "", regex=True)
         .str.replace(r"\r|\n|\t", "", regex=True)
         .str.strip()
         .str.replace(r"\.0$", "", regex=True)
    )

    s = s.replace({
        "": pd.NA,
        "nan": pd.NA,
        "None": pd.NA,
        "<NA>": pd.NA
    })

    return s


def build_lookup(df, key_col, value_col, normalize_value=True):
    temp = df[[key_col, value_col]].copy()
    temp[key_col] = normalize_key(temp[key_col])

    if normalize_value:
        temp[value_col] = normalize_key(temp[value_col])

    temp = temp.dropna(subset=[key_col])
    temp = temp.drop_duplicates(subset=[key_col], keep="first")

    return temp.set_index(key_col)[value_col]


def insert_or_replace_after(df, base_col, new_col, values):
    if new_col in df.columns:
        df[new_col] = values
    else:
        pos = df.columns.get_loc(base_col) + 1
        df.insert(pos, new_col, values)


# -------------------------------------------------
# 1. Clean source key in asset_df
# -------------------------------------------------
asset_puid_clean = normalize_key(asset_df["SubscriptionProductPuid"])


# -------------------------------------------------
# 2. Build lookups
# -------------------------------------------------

# line_df: SubscriptionProductPuid -> ShipToPartySiteId
line_puid_to_site = build_lookup(
    line_df,
    "SubscriptionProductPuid",
    "ShipToPartySiteId",
    normalize_value=True
)

# asset_report_df: CUSTOMER_SITE_ID -> ASSET_ID
site_to_asset_id = build_lookup(
    asset_report_df,
    "CUSTOMER_SITE_ID",
    "ASSET_ID",
    normalize_value=True
)


# -------------------------------------------------
# 3. Map values
# -------------------------------------------------

# asset_df.SubscriptionProductPuid -> line_df.ShipToPartySiteId
shipto_site_raw = asset_puid_clean.map(line_puid_to_site)

# ShipToPartySiteId -> asset_report_df.ASSET_ID
asset_id_raw = shipto_site_raw.map(site_to_asset_id)


# -------------------------------------------------
# 4. Fill not founds
# -------------------------------------------------
shipto_site_final = shipto_site_raw.fillna("#N/A")
asset_id_final = asset_id_raw.fillna("#N/A")


# -------------------------------------------------
# 5. In-progress output
# -------------------------------------------------
asset_inprogress = asset_df.copy()

insert_or_replace_after(
    asset_inprogress,
    "SubscriptionProductPuid",
    "ShipToPartySiteId_FromLine",
    shipto_site_final
)

insert_or_replace_after(
    asset_inprogress,
    "ShipToPartySiteId_FromLine",
    "ASSET_ID_FromAssetReport",
    asset_id_final
)


# -------------------------------------------------
# 6. Transformed output
# -------------------------------------------------
asset_transformed = asset_df.copy()

insert_or_replace_after(
    asset_transformed,
    "SubscriptionProductPuid",
    "ShipToPartySiteId",
    shipto_site_final
)

insert_or_replace_after(
    asset_transformed,
    "ShipToPartySiteId",
    "ASSET_ID",
    asset_id_final
)


# -------------------------------------------------
# 7. Stats
# -------------------------------------------------
stats_df = pd.DataFrame([
    {
        "Field": "ShipToPartySiteId",
        "Total Count": int(asset_puid_clean.notna().sum()),
        "Count Not matching": int((asset_puid_clean.notna() & shipto_site_raw.isna()).sum())
    },
    {
        "Field": "ASSET_ID",
        "Total Count": int(shipto_site_raw.notna().sum()),
        "Count Not matching": int((shipto_site_raw.notna() & asset_id_raw.isna()).sum())
    }
])


# -------------------------------------------------
# 8. Save files
# -------------------------------------------------
with pd.ExcelWriter("asset_inprogress.xlsx", engine="openpyxl") as writer:
    asset_inprogress.to_excel(writer, sheet_name="asset_inprogress", index=False)
    stats_df.to_excel(writer, sheet_name="stats", index=False)

asset_transformed.to_excel("asset_transformed.xlsx", index=False)


# -------------------------------------------------
# 9. Validation prints
# -------------------------------------------------
print("asset_inprogress.xlsx created")
print("asset_transformed.xlsx created")


```
